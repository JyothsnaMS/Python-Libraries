```python
#displot
#jointplot->bivariate analysis
#pairplot->more than 2 independent features
```

Correlation with Heatmap

```python
#colored cells to show 2D correlation matrix
```

```python
df.corr()    #values are only integer or floating points. value ranges between -1 and +1
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())   #required for feature selections
```

```python
#joinplot

sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

```python
#pair plot also known as scatter plot

sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='smoker')
```

```python
#dist plot

sns.displot(df['tip'], kde=True)
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

```python
#categorical plots - box plot, violin plot, count plot, bar plot

sns.countplot('sex', data=df)
```

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
sns.barplot(y='total_bill', x='smoker', data=df)
```

```python
sns.barplot(y='total_bill', x='sex', data=df)
```

```python
sns.boxplot('smoker', 'total_bill', data=df)
```

```python
sns.boxplot('sex', 'total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
sns.boxplot(data=df, orient='h')
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

```python
#violin plot - distribution in terms of kernel density estimation and the box plot
```

```python
sns.violinplot(x='total_bill', y='day', data=df, paletter='rainbow')
```
